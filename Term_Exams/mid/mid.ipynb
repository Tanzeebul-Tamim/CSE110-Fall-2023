### Question 1 [CO6] [10 Points]

Suppose you are a part of a football team that took part in `N` number of football games. Write a Python program that calculates your team's **total points** and **winning percentage** based on the scores from the games. 

First, the program should take the number of games played as input from the user. The scores of the games should also be taken as input from the user in the following format:  
`Your team's score-Opponent team's score`.

For each win, draw, and loss, your team will receive `3`, `1`, and `0` points, respectively. The program should then calculate your team's total points and the winning percentage.

You may assume the number of goals scored in each game will always be less than `10`.

#### Sample Input 1:
```
4
5-2
3-3
1-4
2-1
```

#### Sample Output 1:
```
Total points: 7
Winning percentage: 50.0%
```

#### Explanation:
The first input from the user is `4`, which means that 4 games were played.  
Your team won 2 games (game 1 & 4), drew 1 game (game 2), and lost 1 game (game 3).  
Therefore, total points:  
`3*2 + 1*1 + 0*1 = 7`  
Winning percentage:  
`2/4*100 = 50.0%`

---

#### Sample Input 2:
```
6
2-1
1-0
4-4
3-2
0-3
3-2
```

#### Sample Output 2:
```
Total points: 13
Winning percentage: 66.67%
```

#### Explanation:
The first input from the user is `6`, which means that 6 games were played.  
Your team won 4 games (game 1, 2, 4 & 6), drew 1 game (game 3), and lost 1 game (game 5).  
Therefore, total points:  
`3*4 + 1*1 + 0*1 = 13`  
Winning percentage:  
`4/6*100 = 66.67%`
```

````python
n = int(input("Enter the number of games: "))
total_points = 0
total_wins = 0

for i in range(n):
    score = input(f"Enter the result of game {i + 1} (your_score-opponent_score): ")
    dash_index = 0
    for j in range(len(score)):
        if score[j] == "-":
            dash_index = j
            break
    your_score = int(score[:dash_index])
    opponent_score = int(score[dash_index + 1:])
    
    if your_score > opponent_score:
        total_points += 3
        total_wins += 1
    elif your_score == opponent_score:
        total_points += 1

winning_percentage = (total_wins / n) * 100
print(f"Total points: {total_points}")
print(f"Winning percentage: {round(winning_percentage, 2)}%")
````
