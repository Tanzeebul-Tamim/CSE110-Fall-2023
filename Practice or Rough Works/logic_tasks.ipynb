### Question:

Write a Python program that determines whether a given sequence of integers is a "UB Jumper." A UB Jumper sequence has the following property: for a sequence of length \( n \), the absolute differences between consecutive elements form a set of values that includes every integer from 1 to \( n-1 \) exactly once.

The program should accept input from the user as a space-separated sequence of integers, and it should repeatedly ask for input until the user types "STOP". The program will then check whether the sequence is a UB Jumper or not, and output "UB Jumper" if the sequence meets the criteria, or "Not UB Jumper" otherwise.

- If the input contains non-integer characters, raise a `ValueError` stating "The sequence must contain numbers only."
- If the input consists of only one number, raise a `ValueError` stating "The sequence must contain more than one number."

#### Example Input/Output:

```
Please enter a number sequence separated by space, or 'STOP' to end: 5 1 4 2 3
UB Jumper

Please enter a number sequence separated by space, or 'STOP' to end: 1 4 2 -1 6
Not UB Jumper

Please enter a number sequence separated by space, or 'STOP' to end: 8
ValueError: The sequence must contain more than one number

Please enter a number sequence separated by space, or 'STOP' to end: a b c
ValueError: The sequence must contain numbers only
```

### Instructions:
- Implement the program to handle multiple inputs.
- Check the absolute differences between consecutive numbers to determine whether the sequence is a UB Jumper.


````python
while True:
    sequence = []
    element = ""
    user_input = input(
        "Please enter a number sequence separated by space, or 'STOP' to end: "
    )
    if user_input.upper() == "STOP":
        break

    for i in range(len(user_input)):
        char = user_input[i]
        if ("0" <= char <= "9" or char == "-") and char != chr(32):
            element += char
        elif char == chr(32):
            sequence.append(int(element))
            element = ""
        else:
            raise ValueError("The sequence must contain numbers only")

    if not chr(32) in user_input:
        raise ValueError("The sequence must contain more than one number")

    sequence.append(int(element))

    abs_sequence = []
    ref_seq = []
    stored_for_abs = 0
    for j in range(len(sequence)):
        char = sequence[j]
        if j == 0:
            stored_for_abs = char
        else:
            ref_seq.append(j)
            abs_diff = stored_for_abs - char
            if abs_diff < 0:
                abs_diff *= -1
            abs_sequence.append(abs_diff)
            stored_for_abs = char

    def check_UB_Jumper():
        for k in range(len(abs_sequence)):
            found = False
            for l in range(len(ref_seq)):
                if abs_sequence[k] == ref_seq[l]:
                    found = True
                    break
            if not found:
                return False
        return True

    result = check_UB_Jumper()
    if result:
        print("UB Jumper")
    else:
        print("Not UB Jumper")
````

### Question:

Write a Python program that sorts a given alphanumeric string based on the following criteria:
1. All lowercase letters (`a-z`) are sorted and appear first.
2. All uppercase letters (`A-Z`) are sorted and appear next.
3. All odd digits (`1, 3, 5, 7, 9`) are sorted and appear next.
4. All even digits (`0, 2, 4, 6, 8`) are sorted and appear last.

Implement the program in such a way that the different character types (lowercase letters, uppercase letters, odd digits, and even digits) are stored in separate lists. Sort these lists individually and then combine them in the order mentioned above.

#### Example Input/Output:

**Example 1:**

```
Input: "Bracu1234"
Output: "acruB13524"
```

**Example 2:**

```
Input: "PyThOn2023"
Output: "hnopyTO23120"
```

### Instructions:
- Implement a function `sort_alpha_numerics` that accepts an alphanumeric string and returns the sorted string based on the criteria outlined above.
- The program must handle both uppercase and lowercase letters and digit characters.
- The sorting for each category must be done manually, using a selection sort approach.

````python
def sort_alpha_numerics(alpha_num):
    lowers = []
    uppers = []
    even_digits = []
    odd_digits = []

    for i in range(len(alpha_num)):
        char = alpha_num[i]
        ascii_value = ord(char)
        if 97 <= ascii_value <= 122:
            lowers.append(char)
        elif 65 <= ascii_value <= 90:
            uppers.append(char)
        elif 48 <= ascii_value <= 57:
            if ascii_value % 2 == 0:
                even_digits.append(char)
            else:
                odd_digits.append(char)


    def sort_characters(list):
        for j in range(len(list)):
            min_index = j
            for k in range(j + 1, len(list)):
                if ord(list[k]) < ord(list[min_index]):
                    min_index = k
            temp = list[j]
            list[j] = list[min_index]
            list[min_index] = temp

        result_string = ""
        for l in range(len(list)):
            result_string += list[l]

        return result_string


    sorted_lowers = sort_characters(lowers)
    sorted_uppers = sort_characters(uppers)
    sorted_odd_digits = sort_characters(odd_digits)
    sorted_even_digits = sort_characters(even_digits)
    result = sorted_lowers + sorted_uppers + sorted_odd_digits + sorted_even_digits
    
    return result

result = sort_alpha_numerics("Bracu1234")
print(result)
````

### Question:

Write a Python program that determines the maximum number of teams that can be formed from a group of students based on their previous participation in events.

- The program should first ask the user to input two space-separated integers:
  1. `n`: the total number of students.
  2. `k`: the number of additional events each student needs to participate in.
  
- Next, it should ask the user to input a space-separated sequence of `n` integers, each representing the number of times a particular student has already participated in previous events.

**Rules**:
- A team consists of exactly 3 students.
- A student can only be a part of a team if their total number of participations after the new events will not exceed 5. In other words, for each student with `p` participations, the student can be part of a team if `5 - p >= k`.
  
The program should calculate and return the maximum number of teams that can be formed under these conditions.

#### Example Input/Output:

**Example 1:**

```
Input:
Enter the number of total students and how many times to participate: 6 2
Enter the numbers of times each students have participated previously: 0 4 5 1 2 1

Output:
1
```

**Example 2:**

```
Input:
Enter the number of total students and how many times to participate: 9 1
Enter the numbers of times each students have participated previously: 0 2 3 1 4 3 5 2 1

Output:
2
```

### Instructions:
- Ensure that the input contains only integers.
- Raise a `ValueError` if the input format is incorrect.
- Use the criteria provided to check whether each student is eligible for a team.
- Calculate and output the maximum number of teams that can be formed.

````python
def max_teams(n_and_k, y):
    n_and_k_string = ""
    n_and_k_list = []
    for i in range(len(n_and_k)):
        char = n_and_k[i]
        ascii_value = ord(char)
        if 48 <= ascii_value <= 57 or ascii_value == 32:
            if 48 <= ascii_value <= 57:
                n_and_k_string += char
            elif ascii_value == 32:
                n_and_k_list.append(int(n_and_k_string))
                n_and_k_string = ""
        else:
            raise ValueError("Must contain integers only")
    n_and_k_list.append(int(n_and_k_string))
    if len(n_and_k_list) != 2:
        raise ValueError("Must contain 2 integers")
    
    y_string = ""
    y_list = []
    for j in range(len(y)):
        char = y[j]
        ascii_value = ord(char)
        if 48 <= ascii_value <= 57 or ascii_value == 32:
            if 48 <= ascii_value <= 57:
                y_string += char
            elif ascii_value == 32:
                y_list.append(int(y_string))
                y_string = ""
        else:
            raise ValueError("Must contain integers only")
    y_list.append(int(y_string))
    if len(y_list) != n_and_k_list[0]:
        raise ValueError(f"Must contain {n_and_k_list[0]} integers")
    
    available_participants = []
    for k in range(len(y_list)):
        if 5 - y_list[k] >= n_and_k_list[1]:
            available_participants.append(y_list[k])
    
    max_num_of_teams = len(available_participants) // 3
    return max_num_of_teams

    


user_input_1 = input(
    "Enter the number of total students and how many times to participate: "
)
user_input_2 = input(
    "Enter the numbers of times each students have participated previously: "
)

result = max_teams(user_input_1, user_input_2)
print(result)
````
